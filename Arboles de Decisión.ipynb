### Librerias a Utilizar

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn. tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

### lectura de archivo y visualización de las primeras 5 lineas

```python
#Cargamos el dataset y mostramos las 5 primeras filas
data = pd.read_csv("winequality-red.csv")
data.head()
```

### Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']])
plt.xticks(rotation=40)
plt. title('Boxplot')
plt.ylabel('Valores')
plt.show()
```

### Predictores con la Variable Objetivo

```python
predictors_col = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
predictors
```

```python
target
```

### Correlaciones

```python
corr = predictors.corr()
corr
```

### Matriz de Correlacion

```python
sns.heatmap(corr)
plt.title('Matriz de Correlacion')
plt.show()
```

### Predistores y Target (Traint-Test)

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

### Arbol de Desiciones

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

### Predicciones

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

### Matriz de confusión

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

### Metricas de Desempeño

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, predicciones, average='weighted', zero_division=1)
precision
```

```python
recall = recall_score(y_test, predicciones, average='weighted')
recall
```

```python
f1 = f1_score(y_test, predicciones, average='weighted')
f1
```
